**Analyze if there is a relationship between the price range and the availability of online delivery and table booking**

```python
kpi = df[["Restaurant Name", "Price range", "Has Online delivery", "Has Table booking"]]
kpi
```

```python
print("Price Range Options:", kpi["Price range"].unique())
print("Online Delivery Options:", kpi["Has Online delivery"].unique())
print("Table Booking Options:", kpi["Has Table booking"].unique())
```

```python
online_delivery = kpi.groupby("Price range")["Has Online delivery"].value_counts(normalize = True).unstack() * 100
print(f"Online Delivery Availability by Price Range (%):")
print(online_delivery)

table_booking = kpi.groupby("Price range")["Has Table booking"].value_counts(normalize = True).unstack() * 100
print(f"Table Booking Availability by Price Range (%):")
print(table_booking)
```

```python
online_delivery.plot(kind = "bar", stacked = True, color = ["skyblue", "orange"])
plt.title("Online Delivery availability by Price Range")
plt.xlabel("Price Range")
plt.ylabel("Percentage")
plt.legend("Has Online delivery", loc = "upper right")
plt.grid(axis = "y", linestyle = "--", alpha = 0.7)
plt.show()

table_booking.plot(kind = "bar", stacked = True, color = ["lightgreen", "pink"])
plt.title("Table Booking availability by Price Range")
plt.xlabel("Price Range")
plt.ylabel("Percentage")
plt.legend("Has Table booking", loc = "upper right")
plt.grid(axis = "y", linestyle = "--", alpha = 0.7)
plt.show()
```

```python
online_delivery_ct = pd.crosstab(kpi["Price range"], kpi["Has Online delivery"])
print("\nContingency Table - Online Delivery:")
print(online_delivery_ct )

table_booking_ct = pd.crosstab(kpi["Price range"], kpi["Has Table booking"])
print("\nContingency Table - Table Booking:")
print(table_booking_ct )
```

```python
from scipy.stats import chi2_contingency

chi2_online, p_online, dof_online, expected_online = chi2_contingency(online_delivery_ct)
print("\nChi2 square test for Online Delivery:")
print(f"Chi2_statistics: {chi2_online} p_value: {p_online}")

chi2_table, p_table, dof_table, expected_table = chi2_contingency(table_booking_ct)
print("\nChi2 square test for Table Booking:")
print(f"Chi2_statistics: {chi2_table} p_value: {p_table}")
```

**Determine if higher-priced restaurants are more likely to offer these services**

```python
kpi = kpi[["Restaurant Name", "Has Online delivery", "Has Table booking", "Price range"]]
kpi
```

```python
#Define bins and labels
bins = [0, 2, 4, float('inf')]
labels = ["Low", "Medium", "High"]   #0-1(Low), 2-3(Medium), 4 and above(High)

kpi['category'] = pd.cut(kpi['Price range'], bins = bins, labels = labels, right = False)
print(kpi)
```

```python
price_order = ["Low", "Medium", "High"]
kpi["category"] = pd.Categorical(kpi["category"], categories = price_order, ordered = True)
kpi
```

```python
online_delivery = kpi.groupby("Price range")["Has Online delivery"].value_counts(normalize = True).unstack() *100

table_booking = kpi.groupby("Price range")["Has Table booking"].value_counts(normalize = True).unstack() *100

print("Online delivery Availability by Price Range(%):")
print(online_delivery)

print("Table Booking Availability by Price Range(%):")
print(table_booking)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
kpi["Online_Delivery_Encoded"] = encoder.fit_transform(kpi["Has Online delivery"])
kpi["Table_Booking_Encoded"] = encoder.fit_transform(kpi["Has Table booking"])

x = pd.get_dummies(kpi["Price range"], drop_first = True)
y_online = kpi["Online_Delivery_Encoded"]
model_online = LogisticRegression().fit(x, y_online)
print("Online Delivery logistic Regression Coefficient:", model_online.coef_)

y_table = kpi["Table_Booking_Encoded"]
model_table = LogisticRegression().fit(x, y_table)
print("Table booking Logistic Regression Coeffiecient:", model_table.coef_)
```
