**Task 4 - Online Delivery**

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Dataset2 .csv')
df
```

**Determine the percentage of restaurants that offer online delivery**

```python
kpi = df[['Restaurant Name', 'Has Online delivery']]
kpi
```

```python
online_count =len(kpi[kpi['Has Online delivery'] =="Yes"])
total_restaurants = len(kpi)
 
online_percentage = (online_count / total_restaurants) * 100
print(f"Percentage of restaurants ofering online delivery: {online_percentage:.2f}%.")
```

**Compare the average ratings of restaurants with and without online delivery**

```python
kpi = df[['Restaurant Name', 'Has Online delivery', 'Aggregate rating']]
kpi
```

```python
kpi["Aggregate rating"] = pd.to_numeric(kpi["Aggregate rating"], errors = "coerce")

avg_with_delivery = kpi[kpi["Has Online delivery"] == "Yes"]["Aggregate rating"].mean()
avg_without_delivery = kpi[kpi["Has Online delivery"] == "No"]["Aggregate rating"].mean()

print(f"Average rating (with online delivery) : {avg_with_delivery :.2f}")
print(f"Average rating (without online delivery) : {avg_without_delivery :.2f}")
```
