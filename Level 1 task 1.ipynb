**TASK1 - Top Cuisines**

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Dataset2 .csv')
df
```

```python
df.isnull().sum()
```

```python
df.fillna(value = 'Null')
```

**Determine the top three most common cuisine**

```python
kpi = df[['Cuisines', 'Restaurant Name']]
kpi
```

```python
cuisine_counts = kpi['Cuisines'].value_counts().reset_index()
cuisine_counts.column = ['Cuisines', 'Restaurant_count']
Top_3_cuisines = cuisine_counts.head(3)
print(Top_3_cuisines)
```

```python
plt.subplot(1, 2, 1)
sns.barplot(data = Top_3_cuisines, x = 'Cuisines', y = 'count', palette = 'coolwarm', hue = 'Cuisines', legend = False)
plt.title('Top three most common cuisines', fontsize = 14)
plt.xlabel('Cuisines', fontsize = 12)
plt.ylabel('count', fontsize = 12)
for index, value in enumerate(Top_3_cuisines['count']):
    plt.text(index, value + 5, str(value), ha = 'center', fontsize = 10)
plt.tight_layout()
plt.show()
```

**Determine the percentage of restaurants that serve each of the top cuisines**

```python
cuisine_counts['percentage'] = (cuisine_counts['count'] / cuisine_counts['count'].sum()) * 100
top_percentage = cuisine_counts.head(3)
print(top_percentage)
```

```python
plt.subplot(1, 2, 2)
plt.pie(top_percentage['percentage'], labels = top_percentage['Cuisines'], autopct = '%1.1f', 
        colors = sns.color_palette('coolwarm', len(top_percentage)))
plt.title('Percentage Distribution of Top Cuisines', fontsize = 14)
plt.tight_layout
plt.show()
```
